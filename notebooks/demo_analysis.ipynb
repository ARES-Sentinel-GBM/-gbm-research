## 1. Flux Analysis

Compare metabolic fluxes between GBM and normal astrocytes to identify differential reactions.

```python
from scripts.flux_analysis import run_flux_analysis

# Run flux analysis (demo mode without real model)
# For real analysis, provide actual expression files and Human-GEM model

results_flux = run_flux_analysis(
    gbm_expr_path='data/expression_example.csv',
    astro_expr_path='data/expression_example.csv',
    high_percentile=75,
    low_percentile=25,
    delta_threshold=0.01,
    output_dir='results',
    verbose=True
)
```

```python
# View top results
if len(results_flux) > 0:
    print("Top 10 Differential Reactions:")
    display(results_flux.head(10))
    
    # Plot
    fig = px.bar(
        results_flux.head(20).reset_index(),
        x='abs_delta',
        y='index',
        orientation='h',
        title='Top 20 Differential Fluxes',
        labels={'abs_delta': '|Δ Flux|', 'index': 'Reaction'}
    )
    fig.show()
```

---
## 2. Gene Knockout Analysis

Simulate single and combinatorial gene knockouts to identify potential drug targets.

```python
from scripts.gene_ko import run_gene_knockout

# Define genes of interest (GBM drug targets)
genes = ['RRM1', 'RRM2', 'TYMS', 'GLS', 'LDHA', 'HK2', 'G6PD', 'IDH1']

# Define combinations for synthetic lethality
combos = ['LDHA+RRM2', 'GLS+RRM2', 'HK2+G6PD', 'LDHA+GLS']

# Run knockout analysis
results_ko = run_gene_knockout(
    genes=genes,
    combos=combos,
    output_dir='results',
    verbose=True
)
```

```python
# View results
if len(results_ko) > 0:
    print("Knockout Results Summary:")
    display(results_ko)
    
    # Plot
    fig = px.bar(
        results_ko.sort_values('ratio'),
        x='ratio',
        y='genes',
        orientation='h',
        color='ratio',
        color_continuous_scale='RdYlGn_r',
        title='Gene Knockout Effects on Cell Growth',
        labels={'ratio': 'Growth Ratio (KO/WT)', 'genes': 'Gene Target'}
    )
    fig.add_vline(x=0.1, line_dash='dash', annotation_text='Lethal')
    fig.show()
```
